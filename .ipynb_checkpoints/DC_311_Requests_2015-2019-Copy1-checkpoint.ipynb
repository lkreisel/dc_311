# Introduction

City residents make 311 requests to ask the city to perform a variety of non-emergency services including bulk item collection, missed trash pickup, fallen tree removal, parking enforcement, pothole repair, graffiti removal, and many more. This project looks at all 311
requests made in the city of Washington, DC from 2015 to 2019 in an attempt to find patterns at the Census tract level. Distinct patterns are uncovered that could allow the city to take preemptive action at specific locations and to appropriately distribute resources
where they are needed most. 

This project also explores the relationship between the median household income of a Census tract and 311 ticket response time. Linear regression models found that Census tract median household income can significantly predict ticket response time, controlling for type of request.

The data for this project comes from Washington, DCâ€™s [Open Data DC website](https://opendata.dc.gov/search?q=311). I pulled 311 requests from the years 2015-2019. These years were chosen because they correlate with the most recent American Community Survey (ACS) median household income by Census tract

# Read in Data and Convert to Census Tract

```python
# read in 311 data
d_2015 = pd.read_csv("data/311_City_Service_Requests_in_2015.csv")
d_2016 = pd.read_csv("data/311_City_Service_Requests_in_2016.csv")
d_2017 = pd.read_csv("data/311_City_Service_Requests_in_2017.csv")
d_2018 = pd.read_csv("data/311_City_Service_Requests_in_2018.csv")
d_2019 = pd.read_csv("data/311_City_Service_Requests_in_2019.csv")
```

```python
# read in ACS income data
income = pd.read_csv("data/ACS_Economic_Characteristics_DC_Census_Tract.csv")
```

```python
# read in Census tract locations
ids = pd.read_csv("data/2019_Gaz_tracts_national.txt", sep="\t")

# reduce Census tract data to only show DC tracts
ids_dc = ids[ids['USPS'] == "DC"]

# remove whitespace from column names
ids_dc.columns = ids_dc.columns.str.replace(' ', '')
```

Every 311 request is associated with a latitude and longitude. To match these latitudes and longitudes to Census tracts, let's calculate the spatial distance between the latitude and longitude for each service request and the latitude and longitude of each of the 179 Census tracts as listed in the Census files on the US Census Bureau website. Then we'll find the smallest distance and assign the request to the associated Census tract. 

```python
# function to get the Census tract of each service request

def get_census_tract(data_df, tract_df): 

    # get geopandas df from data_df
    gdf1 = geopandas.GeoDataFrame(
        data_df, geometry=geopandas.points_from_xy(data_df['LATITUDE'], 
                                              data_df['LONGITUDE']))
    # get geopandas df from tract_df
    gdf2 = geopandas.GeoDataFrame(
        tract_df, geometry=geopandas.points_from_xy(tract_df['INTPTLAT'], 
                                                  tract_df['INTPTLONG']))
    # set buffer for merge
    gdf2['geometry'] = gdf2.geometry.buffer(0.01)

    # geopandas spatial join
    df = gdf1.sjoin(gdf2, how="left")
    
    # remove values with no match given buffer
    df = df[df['INTPTLAT'].notna()]
    df = df[df['INTPTLONG'].notna()]

    # get tuples of coordinate pairs
    df['coords'] = list(zip(df.LATITUDE, df.LONGITUDE))
    df['coords_tract'] = list(zip(df.INTPTLAT, df.INTPTLONG))

    # calculate distance between the coordinates
    df['distance'] = df.apply(lambda x: distance.distance(x['coords'], x['coords_tract']), axis=1)

    # sort values by distance and take the smallest distance
    df = df.sort_values('distance').groupby('OBJECTID', as_index=False).first()
    
    return df
```

```python
# run function for each year's data

a_2015 = get_census_tract(d_2015, ids_dc)
a_2016 = get_census_tract(d_2016, ids_dc)
a_2017 = get_census_tract(d_2017, ids_dc)
a_2018 = get_census_tract(d_2018, ids_dc)
a_2019 = get_census_tract(d_2019, ids_dc)

# concat all years together
all_data = pd.concat([a_2015, a_2016, a_2017, a_2018, a_2019])
```

```python
# merge with income data 
data = all_data.merge(income, on='GEOID', how = 'left')
```

## Find proportion of requests by Ward

```python
# find what proportion of requests from every ward
sumsByWard = data.groupby(['WARD']).agg({'WARD': "count"}).rename(columns={'WARD':'TotalReqs'})
sbw = sumsByWard.reset_index()
sbw['WARD'] = sbw['WARD'].astype(int)
sbw["proportion"] = round(sbw['TotalReqs'] / sum(sbw['TotalReqs']) * 100, 2)
sbw['WARD'] = sbw['WARD'].astype(int)
sbw['proportion'] = sbw['proportion'].astype(str)
```

```python
#### import warnings
warnings.filterwarnings('ignore')

# show proportions on DC map with ward boundaries
w = geopandas.read_file("data/Wards_from_2022.shp")
w = w.to_crs("EPSG:4326")

w_x = w.merge(sbw, left_on="WARD", right_on="WARD", how="left")

fig, ax = plt.subplots(figsize=(10,10))

w_x.plot(ax=ax)
w_x.apply(lambda x: ax.annotate(
    s=x.NAME + "\n" + x.proportion + "%", 
    xy=x.geometry.centroid.coords[0],
    ha='center', 
    fontsize=10, color='white'
),axis=1);
ax.set_axis_off()
plt.title("Proportion of 311 Requests by Ward (2015-2019)", fontsize=20)
```

Wards are approximately equal in population. Though Ward 2 makes up approximately 12.5% the population of DC, the residents and visitors of Ward 2 are generating about 19% of all 311 requests. Though Wards 3 and 8 make up approximately 12.5% of DC's population each, the residents and visitors of Wards 3 and 8 are generating 9% and 8% of DC's 311 requests. 

# Find proportions of service requests by ward

```python
dataWard = data.groupby(['SERVICECODEDESCRIPTION', 'WARD']).agg({'SERVICECODEDESCRIPTION': "count"}).rename(columns={'SERVICECODEDESCRIPTION':'Count'}).reset_index()
wards = dataWard.merge(sumsByWard,on='WARD',how='left').rename(columns={'SERVICECODEDESCRIPTION':'Request_Type'})
wards['Proportion'] = wards['Count'] / wards['TotalReqs'] * 100
```

First, let's calculate the proportion of each request. For example, 16% of all requests from Ward 1 were "Parking Enforcement." Then, let's look at the service requests that have the biggest disparities in proportions between wards. The figure below shows the requests that had the biggest disparities bretween the maximum proportion from a ward and the minimum proportion from a different ward. 

```python
# summarize at service code level; find diff between max and min
diffs = wards.groupby('Request_Type')['Proportion'].agg(np.ptp).sort_values(ascending=False)
diffs = diffs[diffs>1]

# reduce data set to only these request types
wardsRe = wards.loc[wards['Request_Type'].isin(diffs.index)]
```

```python
wardsRe['WARD'] = wardsRe['WARD'].astype(int)
h = seaborn.catplot(x="Proportion", y="Request_Type", jitter=False, hue='WARD', data=wardsRe).set(title='Proportion of Service Requests by Ward')
h.fig.set_figwidth(14)
h.fig.set_figheight(10)
```

As the figure shows, Ward 2 is making a disproportionately high number of requests for parking meter repairs compared to every other ward.Over 31% of Ward 1 requests and over 29% of Ward 2 requests involve parking (parking enforcement, parking meter repair, and emergeny no-parking verification). # Map Requests by Census Tract

```python
# load Census tract shape files
df = geopandas.read_file("tl_2019_11_tract/tl_2019_11_tract.shp")
df = df.to_crs("EPSG:4326")
df['GEOID'] = df['GEOID'].astype(int)
```

```python
# function to plot request count by Census tract

def plot_requests(request):

    service = data[data['SERVICECODEDESCRIPTION'] == request]
    serviceByGeo = service.groupby(['GEOID']).agg({'NAME': "sum"}).rename(columns={'NAME':'TotalReqs'}).reset_index()
    serviceByGeo['GEOID'] = serviceByGeo['GEOID'].astype(int)
    map_and_stats = df.merge(serviceByGeo, left_on="GEOID", right_on="GEOID", how="left")
    map_and_stats['TotalReqs'] = map_and_stats['TotalReqs'].fillna(0)

    fig, ax = plt.subplots(1, figsize=(8, 8))
    map_and_stats.plot(column="TotalReqs", cmap="Reds", linewidth=0.4, ax=ax, edgecolor=".4")
    plt.title("Washington, DC 311 Requests by Census Tract: \n " + request + " (2015-2019)", fontsize=15)
    bar_info = plt.cm.ScalarMappable(cmap="Reds", norm=plt.Normalize(vmin=min(map_and_stats['TotalReqs']), vmax=max(map_and_stats['TotalReqs'])))
    bar_info._A = []
    cbar = fig.colorbar(bar_info)
    ax.set_axis_off()
```

Wards 1 and 2 are the most densely populated areas of the city and the areas most often frequented by tourists so this pattern of parking requests is not surprising.

```python
# calculate time between service request and service resolution
data['timeToComplete'] = data['RESOLUTIONDATE'] - data['SERVICEORDERDATE']
data['timeToComplete'] = data['timeToComplete'].fillna(pd.Timedelta(seconds=0))
```

Let's remove requests that had a completion rate of less than 2 hours (thinking this must be an error or a rejected request that was marked as completed) and greater than 720 hours (30 days).

```python
# bulk collection 
bulk = data[data['SERVICECODEDESCRIPTION'] == "Bulk Collection"]
bulk['timeToComplete'] = bulk['timeToComplete'].dt.total_seconds()
bulk['timeToComplete'] = bulk['timeToComplete'] / 3600
bulk = pd.DataFrame(bulk, columns=['timeToComplete', 'SERVICEORDERSTATUS','DP03_0062E', 'DETAILS', 'WARD','NAME'])

# remove all data where timetocomplete is less than 12 hours
bulk = bulk[bulk['timeToComplete'] > 2]
bulk = bulk[bulk['timeToComplete'] < 720]
bulk = bulk[bulk['DP03_0062E'].notna()]
```

# Conclusions

The 311 data set exhibits distinct patterns by Census tract and by ward. Some of these
patterns can be used by the city to take preventative action. For example, there is a hyper-
specific location that disproportionately makes the majority of illegal dumping requests. The
city could install cameras or no dumping signs in this area. One area of the city is also
producing a very concentrated number of requests for tree inspections. The city may want
send arborists to the area to see if a disease is traveling through the population of trees.
Inspectors may want to frequent the two Census tracts that have high concentrations of illegal
construction requests and check for signs of un-permitted work.

Linear regression models showed that the median household income of a Census tract can
significantly predict ticket response time. The city may want to consider allocating more
resources to the Census tracts with longer response times, but further analysis would be
needed before taking any action based on these models. Further analysis would include trying
to detect any seasonal patterns or other factors that might affect the response times.
